9. Create a structured array representing a position (x,y) and a color (r,g,b).

````python
points = np.random.random((100, 2))
# points 

D = np.zeros((100, 100))  # distances of points point[i, :], point[j, :]
for i in range(100):
  for j in range(100):
    point_1 = points[i, :]
    point_2 = points[j, :]
    D[i, j] = np.sqrt(((point_1 - point_2) ** 2).sum())

print(D)
````
