### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo.head(10)
```

```python
chipo["item_price"] = pd.to_numeric(chipo["item_price"].str.lstrip('$'))
```

```python
pd.to_numeric(chipo["item_price"])
```

```python
chipo["item_price"] = lambda x: x["item_price"].str.lstrip('$')
```

```python
chipo["item_price"] = chipo.assign(Price = lambda x: x["item_price"].str.lstrip('$'))
```

### Step 4. How many products cost more than $10.00?

```python
df = chipo[chipo.item_price > 10]
df.item_name.nunique()
```

### Step 6. Sort by the name of the item

```python
df.sort_values(['item_name','item_price'], ascending = True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
df.sort_values(['item_price'], ascending = False)
```
